```python
# import required packages
import pandas as pd
import numpy as np

from sklearn.feature_extraction.text import CountVectorizer
```

```python
df = pd.read_csv('../../data/train_data.csv')

# dataframe shape: (rows, columns)
display(df.shape)

# first 5 datapoints
display(df.head())
```

```python
# check if null value exist
display(df.clean_sentence.isna().sum())

# replace null values with empty string
df.clean_sentence = df.clean_sentence.fillna('')

# verify null count
display(df.clean_sentence.isna().sum())
```

```python
# instantiate a CountVectorizer
cv = CountVectorizer()

# training and creating bag of words
counts = cv.fit_transform(df['clean_sentence'].values)
```

```python
# extract bag of words as dataframe
df_bow = pd.DataFrame(counts.A, columns=cv.get_feature_names_out())

# include Sentiment column
df_bow['Sentiment'] = df['Sentiment']

# shape
display(df_bow.shape)

# 10 random datapoints
df_bow.sample(10)
```
